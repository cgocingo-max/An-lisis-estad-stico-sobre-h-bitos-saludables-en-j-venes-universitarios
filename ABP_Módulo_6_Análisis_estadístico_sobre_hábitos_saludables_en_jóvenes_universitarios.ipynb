# Lección 4: Distribución muestral y Teorema del Límite Central

## Distribución muestral

La distribución muestral corresponde a la distribución de una estadística calculada a partir de múltiples muestras extraídas de una misma población.

En este estudio se utilizará el conjunto de datos de estudiantes universitarios como población inicial y se generarán múltiples muestras aleatorias para analizar cómo varía la media de ciertas variables.

Esto permite observar cómo se comportan las medias muestrales y analizar su dispersión en relación con la media poblacional.


## Generación de múltiples muestras

```python
import numpy as np

# Generar 1000 muestras de tamaño 50
sample_means = []

for i in range(1000):
    sample = data["study_hours"].sample(n=50)
    sample_means.append(sample.mean())
```

## Distribución muestral de la media

```python
import matplotlib.pyplot as plt

plt.hist(sample_means, bins=30)

plt.title("Distribución muestral de la media de horas de estudio")
plt.xlabel("Media de horas de estudio")
plt.ylabel("Frecuencia")

plt.show()
```

## Comparación entre media poblacional y media muestral

```python
# media poblacional
media_poblacion = data["study_hours"].mean()

# media de las medias muestrales
media_muestral = np.mean(sample_means)

media_poblacion, media_muestral
```

## Verificación empírica del Teorema del Límite Central

Los resultados muestran que la media de las medias muestrales es muy cercana a la media poblacional.

Esto confirma empíricamente el Teorema del Límite Central, el cual establece que al tomar múltiples muestras aleatorias de tamaño suficiente, la distribución de sus medias tiende a aproximarse a una distribución normal y su promedio se aproxima al valor real de la población.

## Efecto del tamaño muestral en la dispersión

Este comportamiento está relacionado con el error estándar de la media, el cual disminuye cuando aumenta el tamaño muestral.

# Lección 5: Inferencia e Intervalos de confianza para la media

# Cálculo de Intervalo de Confianza (Primera Variable)

```python
import numpy as np
import scipy.stats as stats

# Intervalo de confianza para study_hours

media = data["study_hours"].mean()
std = data["study_hours"].std()
n = len(data)

intervalo_95 = stats.t.interval(0.95, df=n-1, loc=media, scale=std/np.sqrt(n))

intervalo_95
```

Con 95% de confianza la media real está en ese rango

# Cálculo de Intervalo de Confianza (Segunda Variable)

```python
# Intervalo de confianza para sleep_hours

media_sleep = data["sleep_hours"].mean()
std_sleep = data["sleep_hours"].std()

intervalo_sleep_95 = stats.t.interval(0.95, df=n-1, loc=media_sleep, scale=std_sleep/np.sqrt(n))

intervalo_sleep_95
```

```python
ic_90 = stats.t.interval(0.90, df=n-1, loc=media, scale=std/np.sqrt(n))
ic_95 = stats.t.interval(0.95, df=n-1, loc=media, scale=std/np.sqrt(n))
ic_99 = stats.t.interval(0.99, df=n-1, loc=media, scale=std/np.sqrt(n))

ic_90, ic_95, ic_99
```

```python
# Tomar muestra pequeña

sample_small = data.sample(50)

media_small = sample_small["study_hours"].mean()
std_small = sample_small["study_hours"].std()
n_small = len(sample_small)

intervalo_small = stats.t.interval(0.95, df=n_small-1, loc=media_small, scale=std_small/np.sqrt(n_small))

intervalo_small
```

Los intervalos de confianza permiten estimar el rango en el cual se encuentra el verdadero valor de la media poblacional.

Los resultados muestran que las medias de variables como las horas de estudio y las horas de sueño se encuentran dentro de rangos relativamente acotados cuando se utilizan muestras grandes.

# Lección 6: Test de Significancia

Test de hipótesis para la media

Primero hacemos una prueba para ver si la media de la muestra es distinta a la media poblacional.

```python
data.columns
```

```python
import scipy.stats as stats

# seleccionamos la variable de interés
exam_scores = data['exam_score']

# media poblacional hipotética
mu = 70

# test t de una muestra
t_stat, p_value = stats.ttest_1samp(exam_scores, mu)

print("Estadístico t:", t_stat)
print("Valor p:", p_value)
```

# Interpretación Test de hipótesis para la media

Se realizó una prueba **t de una muestra** para evaluar si el promedio de los puntajes del examen (`exam_score`) es diferente de 70.

**Hipótesis:**

H₀: μ = 70
H₁: μ ≠ 70

El resultado del test entregó:

* Estadístico t = -298.42
* Valor p = 0.0

Dado que el **valor p es menor que α = 0.05**, se **rechaza la hipótesis nula**.

Esto indica que existe evidencia estadística suficiente para afirmar que el promedio del puntaje de examen en la muestra **es significativamente diferente de 70**.


### Comparación del valor-p con α

Decisión estadística: para tomar la decisión se compara el **valor p** con el nivel de significancia **α = 0.05**.

* Si p < α → se rechaza la hipótesis nula.
* Si p ≥ α → no se rechaza la hipótesis nula.

En este caso:

p = 0.0 < 0.05

Por lo tanto, **se rechaza la hipótesis nula**, lo que indica que la media observada es significativamente diferente del valor hipotético de 70.

### Error tipo I y error tipo II

En el contexto de este análisis:

* **Error tipo I:** rechazar la hipótesis nula cuando en realidad es verdadera.
  En este caso sería concluir que el promedio del puntaje de examen es diferente de 70 cuando en realidad sí es 70.

* **Error tipo II:** no rechazar la hipótesis nula cuando es falsa.
  En este caso sería concluir que el promedio del puntaje de examen es 70 cuando en realidad es diferente.

El nivel de significancia α = 0.05 representa la probabilidad de cometer un **error tipo I**.

### Interpretación

El test de hipótesis realizado muestra que el valor p obtenido es menor que el nivel de significancia α = 0.05. Por lo tanto, se rechaza la hipótesis nula.

Esto indica que existe evidencia estadística suficiente para afirmar que el promedio de los puntajes de examen en el conjunto de datos analizado es significativamente diferente de 70.

A través del análisis exploratorio, intervalos de confianza y pruebas de hipótesis, se observó que las horas de estudio tienen una relación significativa con el puntaje obtenido en los exámenes.

Los intervalos de confianza permitieron estimar con distintos niveles de certeza los valores promedio de variables clave del estudio, mientras que el test de hipótesis confirmó estadísticamente que la media observada difiere de la media poblacional planteada en la hipótesis nula.

Además, los gráficos permitieron visualizar patrones importantes en los datos, como la relación positiva entre horas de estudio y resultados académicos, y posibles efectos negativos asociados al uso excesivo de redes sociales.

En conjunto, estos resultados muestran cómo las herramientas de estadística inferencial permiten comprender mejor los factores que influyen en el rendimiento académico.
